This notebook aims to implement the spectral clustering algorithm, as discussed in:
https://uvle.upd.edu.ph/mod/forum/discuss.php?d=53869

The algorithm computes a graph Laplacian from a matrix of training data, then uses its eigenvectors to perform clustering. For a cluster number of 2, the Fiedler vector is utilized in predicting the class labels. For larger cluster numbers, the eigenvectors are used as features and a simple k-means clustering algorithm is employed (other clustering methods may be used if available already in the package).

```python
import numpy as np
import matplotlib.pyplot as plt

from dilpack.clustering.spectral import SpectralClustering
```

```python
# Plotting function
def visualize_clusters(X, y, title):
    fig, ax = plt.subplots(figsize=(9,7))
    ax.set_title(title, fontsize=18)
    ax.scatter(X[:, 0], X[:, 1],c=y,s=50, cmap="viridis")
```

```python
# Load training set A
X_a = np.load("data/spectral_train_set_a.npy")

# Perform spectral clustering with K=2 (moon-shaped clusters)
pred_a = SpectralClustering(X_a, 2).fit(X_a)

# Plot clusters and labels
visualize_clusters(X_a, pred_a, "Spectral Clustering with Complex Cluster Shape")
```

```python
# Load training set B
X_b = np.load("data/spectral_train_set_b.npy")

# Perform spectral clustering with K=3
pred_b = SpectralClustering(X_b, 3).fit(X_b)

# Plot clusters and labels
visualize_clusters(X_b, pred_b, "Spectral Clustering with Three Clusters")
```
